```python
inputfile.split()
name = inputfile.split()
name = [x for x in name if "!" not in x]  # gets rid of entries with !
name = [x for x in name if "pecies" not in x]  # gets rid of the word species
#print name
#len(name)
```

```python
raw_smiles = name
smiles = {}
for n in range(0, len(name)):
    s = raw_smiles[n]
    s = re.sub('A1-', '[C]1CCCCC1', s)
    s = re.sub('A1', 'C1CCCCC1', s)
    s = re.sub('AR', 'Ar', s)
    s = re.sub('CH3', '[CH3]', s)
    s = re.sub('CH4', 'C', s)
    s = re.sub('C2H ', '[C]#C', s)
    s = re.sub('CH2', '[CH2]', s)
    s = re.sub('CO2', 'O=C=O', s)
    s = re.sub('OH', 'O', s)
    s = re.sub('O1 ', 'O', s)
    s = re.sub('O2H', 'OO', s)
    s = re.sub('O2 ', '[O][O]', s)
    s = re.sub('O2', 'O[O]', s)
    s = re.sub('CHO ', 'C=O', s)
    s = re.sub('CO', 'O=[C]', s)
    s = re.sub('CH2', '[CH2]', s)
    s = re.sub('COCH2', '(=O)CC', s)
    s = re.sub('COCH3 ', 'CC(C)=O', s)
    s = re.sub('C2H2', 'C#C', s)
    s = re.sub('C2H3', '[CH]=C', s)
    s = re.sub('C2H4', 'C=C', s)
    s = re.sub('C2H5', 'C[CH2]', s)
    s = re.sub('A2', 'C1CCC2CCCCC2C1', s)
    s = re.sub('CHOH', '[CH]O', s)
    s = re.sub('HO2', '[O]O', s)
    
    s = re.sub('C2H6', 'CC', s)
    s = re.sub('C3H8', 'CCC', s)
    s = re.sub('-C4H10', 'CC(C)C', s)
    s = re.sub('C4H10', 'CCCC', s)
    
    
    #for x in range(1, 10):  # saturated hydrocarbons
    #    pattern = r'^C' + re.escape(x) + r'H' + re.escape(2*x+2)
    #    s = re.sub(pattern, 'C' * x, s)
    smiles[n] = s
    
a = smiles
a.items()
new_smiles = [(v,k) for (k,v) in a.iteritems()]
new_smiles = zip(*new_smiles)[0]
names = zip(name, new_smiles)
for line in names:
    print(line)
```

```python
#Formatting data into dictionary
names = inputfile.split()
names = [x for x in names if "!" not in x]
names = [x for x in names if "END" not in x]
names = [x for x in names if "DILUENTS" not in x]
names = [x for x in names if "\\" not in x]
inames = map(lambda f: '!' + f + '!', names)

names = dict(zip(names, inames))
```

```python
#Automated smile creation
raw_smiles = {}
for name, iname in names.iteritems():
    s = iname
    s = re.sub('!AR!', '[Ar]', s)
    s = re.sub('!C!', '[C]', s)
    s = re.sub('!CH2!', '[CH2]', s)
    s = re.sub('!CH3!', '[CH3]', s)
    s = re.sub('!CO!', 'CO', s)
    s = re.sub('!CO2!', '[O=C=O]', s)
    s = re.sub('!C2H3!', '[CH]=C', s)
    s = re.sub('!HO2!', '[O]O', s)
    s = re.sub('!O!', '[O]', s)
    s = re.sub('!OH!', '[OH]', s)
    s = re.sub('!O2!', '[O][O]', s)
    s = re.sub('!C2O!', '[C]#C[O]', s)
  
    raw_smiles[name] = s
```

```python
#Testing smiles
successful = {}
unsuccessful = {}
for name, smiles in raw_smiles.iteritems():
    try:
        molecule = Chem.MolFromSmiles(smiles)
        smiles = Chem.MolToSmiles(molecule, True)
        successful[name] = smiles
    except:
        unsuccessful[name] = smiles
        
for name in sorted(successful.keys()):
    smiles = successful[name]
    print "{:20}\t{:20}\t Autoconfirm".format(name, smiles)
for name in sorted(unsuccessful.keys()):
    smiles = unsuccessful[name]
    print "{:20}\t{:20}\t Not Converted".format(name, smiles)
print "{}\tsuccesses out of {}\n{}% success rate".format(len(successful), len(names), round(float(len(successful))/len(names) * 100))
```
